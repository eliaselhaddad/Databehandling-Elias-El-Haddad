```python
missing_values = df.isnull().sum().sort_values(ascending=True)
missing_values = missing_values[missing_values > 0]
missing_values
```

```python
freetime_nan = df[df["freetime"].isnull()].index.tolist()
print(f"These row indices: {freetime_nan}")
```

```python
freetime_age_nan = df[df["freetime"].isnull() | df["age"].isnull()].index.tolist()
print(f"These row indices: {freetime_age_nan}")
```

```python
number_of_rows = df.shape[0]
number_of_columns = df.shape[1]
print(f"Number of rows: {number_of_rows}")
print(f"Number of columns: {number_of_columns}")
rows_missing_multiple_values = df.loc[df.isnull().sum(axis=1) > 1].shape[0] 
print(f"Number of rows with multiple missing values: {rows_missing_multiple_values}")
proportion = rows_missing_multiple_values / number_of_rows
print(f"These rows consitute {proportion:.4f} of the dataset")
```
